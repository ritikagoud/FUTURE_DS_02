```python
import pandas as pd
df = pd.read_csv("Advertising_Data.csv") 
```

```python
total_sales = df["Product_Sold"].sum()
avg_sales = df["Product_Sold"].mean()
correlation = df.corr()["Product_Sold"].sort_values(ascending=False)
print("Total Products Sold:", total_sales)
print("Average Products Sold per Campaign:", round(avg_sales, 2))
print("\nCorrelation with Product_Sold:\n", correlation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x=correlation.index, y=correlation.values, palette="coolwarm")
plt.title("Correlation of Marketing Channels with Product Sales", fontsize=14, weight='bold')
plt.ylabel("Correlation Coefficient")
plt.xlabel("Marketing Channels")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap", fontsize=14, weight='bold')
plt.show()
```

```python
import numpy as np
np.random.seed(42)

df["Impressions"] = np.random.randint(50000, 200000, len(df))
df["Clicks"] = np.random.randint(1000, 10000, len(df))
df["Cost"] = np.random.uniform(1000, 10000, len(df))
```

```python
df["CTR"] = (df["Clicks"] / df["Impressions"]) * 100
df["CPC"] = df["Cost"] / df["Clicks"]
df["ROI"] = (df["Product_Sold"] - df["Cost"]) / df["Cost"]
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Cost', y='ROI', data=df)
plt.title("ROI vs Cost (Campaign Efficiency)")
plt.xlabel("Total Cost")
plt.ylabel("ROI")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['CTR'], kde=True, bins=20)
plt.title("Distribution of Click-Through Rate (CTR)")
plt.xlabel("CTR (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Clicks', y='CPC', data=df)
plt.title("CPC vs Clicks (Cost per Click Efficiency)")
plt.xlabel("Clicks")
plt.ylabel("CPC (Cost per Click)")
plt.show()
```

```python
channel_cols = ['TV','Billboards','Google_Ads','Social_Media','Influencer_Marketing','Affiliate_Marketing']

plt.figure(figsize=(10,6))
sns.barplot(x=channel_cols, y=[df[c].mean() for c in channel_cols])
plt.title("Average Spending on Each Marketing Channel")
plt.xticks(rotation=45)
plt.ylabel("Average Spend")
plt.show()
```
